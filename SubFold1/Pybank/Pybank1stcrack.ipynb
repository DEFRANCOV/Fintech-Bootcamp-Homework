```python
#This script will use the Pathlib library to set the file path,
#use the csv library to read in the file, and iterate over each
#row of the file to calculate customer sales averages.
"""

# @TODO: Import the pathlib and csv library

from pathlib import Path
import csv

# @TODO: Set the file path
csvpath = Path('Sales.csv')

#Your task is to create a Python script that analyzes the records to calculate each of the following:


#The total number of months included in the dataset.


#The net total amount of Profit/Losses over the entire period.


#The average of the changes in Profit/Losses over the entire period.


#The greatest increase in profits (date and amount) over the entire period.


#The greatest decrease in profits (date and amount) over the entire period.


#Your resulting analysis should look similar to the following:
#Financial Analysis
#----------------------------
#Total Months: 86
#Total: $38382578
#Average  Change: $-2315.12
#Greatest Increase in Profits: Feb-2012 ($1926159)
#Greatest Decrease in Profits: Sep-2013 ($-2196167)
#Your final script should print the analysis to the terminal and export a text file with the results.
```

```python
# Import the file using pathlib
from pathlib import Path
import csv
csvpath = Path("PyBank_Resources_budget_data.csv")

#Initialize variables 
numberofmonths = (0)
totalprofitlosses = (0)

#to calculate the change in profit/loss, we need the current profit/loss and the immediate prior profit/loss
monthlyprofitlossestimezero = (0)
monthlyprofitlossestimenegativeone = (0)

#in the for loop, monthlychangeprofits = monthlyprofitlossestimezero - monthlyprofitlossestimenegativeone
monthlychangeprofits = (0)
greatestchangeprofits = (0)
smallestchangeprofits = (0)

#To calculate the average change in profits, we need to calculate the RATIO: (total change in profits/the total number of "changes").
#For any list of numbers with "X" representing the count of numbers in the list, "X-1' will represent the count of changes in the numbers.
#With this in mind, we can't simply accept the value of totalprofitlosses, at the end of the looping process, as the numerator for RATIO.
#This is because the looping process will record the first monthly profit/loss as a change (i.e. from zero to the first value), which is not
#correct for our objective.
#Thus, we need to exclude the very first set of values from the ratio;  So we will record the first change in profit/loss
#and deduct this amount from the numerator which is total cumulative change.   We will also deduct 1 from the denominator, which in
#this case is # ofmonths.
totalmonthlychangeprofits = (0)
firstmonthchangeprofits = (0)
averagechangeprofits = (0)

#Open the csvfile as an object
#Create a header
with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    header = next(csvreader)
    #print(f"{header} <------ HEADER")

# -------------------------------------------------------------
# The iterative process will calculate the following:
    #Total # of months, Cumlative Profit or Loss;
    #Smallest change in profits, Greatest change in profits;
   
    
    for row in csvreader:
        #accumulate the total # of months through each iteration
        numberofmonths += 1  
        
        #convert the value in first column from text to integer, this is the current monthly profit/loss
        monthlyprofitlossestimezero = int(row[1])   
        
        #accumulate the total profits/losses through each iteration
        totalprofitlosses += monthlyprofitlossestimezero
       
        #change in profits/losess = current value - last month's value
        monthlychangeprofits = monthlyprofitlossestimezero - monthlyprofitlossestimenegativeone
        
        #accumulate the total change in profits/losses through each iteration
        totalmonthlychangeprofits += monthlychangeprofits
        
        #for the very first iteration, we must record the change in profit/loss in order to properly calculate
        #the average monthly change in profit/loss.   See comments above initialization of totalmonthlychangeprofits.
        if numberofmonths == 1:
            firstmonthchangeprofits = monthlychangeprofits
            
        #calculate the smallest and greatest change in monthly profit/loss
        if smallestchangeprofits == 0:
            smallestchangeprofits = monthlychangeprofits
        elif monthlychangeprofits < smallestchangeprofits:
            smallestchangeprofits = monthlychangeprofits
        elif monthlychangeprofits > greatestchangeprofits:
            greatestchangeprofits = monthlychangeprofits
        #print(numberofmonths, totalprofitlosses, monthlychangeprofits, smallestchangeprofits, greatestchangeprofits, totalmonthlychangeprofits, row)
        
        #reset the prior monthly profit/loss to current prior monthly profit/loss
        #BECAUSE THE CURRENT VALUE IS THE PRIOR VALUE AT THE NEXT ITERATION
        monthlyprofitlossestimenegativeone = monthlyprofitlossestimezero
    
    #print(firstmonthchangeprofits)
    averagechangeprofits = round((totalmonthlychangeprofits - firstmonthchangeprofits) / (numberofmonths - 1),2)
    #print(averagechangeprofits)
    
print("FINANCIAL ANALYSIS")
print("-" * 40)
print(f"Total Number of Months: {numberofmonths}")
print(f"Total cumulative profit: ${totalprofitlosses}")
print(f"Average change in monthly profit: ${averagechangeprofits}")
print(f"Largest increase in monthly profit:  $({greatestchangeprofits})")
print(f"Largest dencrease in monthly profit:  $({smallestchangeprofits})")
    
# Set the output file path
output_path = Path("pybankoutput.txt")

# Open the output_path as a file object in "write" mode ('w').
# Write the appropriate calculations into the contents of 'pybankoutput'.
with open(output_path, 'w') as file:
    file.write("FINANCIAL ANALYSIS")
    file.write("\n")
    file.write("-" *40)
    file.write("\n")
    file.write(f"Total Number of Months: {numberofmonths}\n")
    file.write(f"Total cumulative profit: ${totalprofitlosses}\n")
    file.write(f"Average change in monthly profit: ${averagechangeprofits}\n")
    file.write(f"Largest dencrease in monthly profit:  $({smallestchangeprofits})\n")
```
